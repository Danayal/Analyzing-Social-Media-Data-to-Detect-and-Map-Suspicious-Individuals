There are a limited number of free sentiment analyzer tools available online. By using a freemium tool like Parralel Dots, a score for fear, sad, bored, happy, excited, and angry can be generated. However, the limitation wtih this freemium tool is its restriction of calling only 1000 API requests a day.

Using an open source free tool like TextBlob allows us to call it as many times as possible, however, it has certain limitations: it is trained on an old IMDb Movies Review dataset and only has two binary continuous output: a high positive score for "polarity" indicates positivity and a high negative score indicates negativity. A high positve score for  the "subjectivity" object indicated subjectivity and a high negative score indicates objectivity.



The limitations of TextBlob's built in classifier can be eradicated by training a new classifier by feeding it a larger annoted data set.

```python
def get_followings(username):

    c = twint.Config()
    c.Username = username
    c.Pandas = True

    twint.run.Following(c)
    list_of_followings = twint.storage.panda.Follow_df

    return list_of_followings['following'][username]
```

```python
followings = {}
following_list = []
for person in users:
    print('#####\nStarting: ' + person + '\n#####')
    try:
        followings[person] = get_followings(person)
        following_list = following_list + followings[person]
    except KeyError:
        print('IndexError')
```

```python
Counter(following_list).most_common()
```

```python
follow_relations ={}
for following_user in followings.keys():
    follow_relation_list = []
    for followed_user in followings.keys():
        if followed_user in followings[following_user]:
            follow_relation_list.append(True)
        else:
            follow_relation_list.append(False)
    follow_relations[following_user] = follow_relation_list
```

```python
following_df = pd.DataFrame.from_dict(follow_relations, orient='index', columns=followings.keys())
```
